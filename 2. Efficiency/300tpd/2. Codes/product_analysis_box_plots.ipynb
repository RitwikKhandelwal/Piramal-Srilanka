```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
seaborn.set(style='whitegrid')
```

```python
#df = pd.read_csv("Flat_File_Forehearth_Level_DPR_Eff_J20.csv")
df = pd.read_csv("../flat file 180ML DCSL LW HIP.csv")
```

```python
df.info()
```

```python
df["Type"] = ""*len(df)
```

```python
c1 = df.index[(df["P5_eff"]>=0) & (df["P5_eff"]<=90.6)]
c2 = df.index[(df["P5_eff"]>=90.6) & (df["P5_eff"]<=91.7)]
c3 = df.index[(df["P5_eff"] >=91.7) & (df["P5_eff"] <=93)]
c4 = df.index[(df["P5_eff"] >= 93) & (df["P5_eff"] <= 99)]

c1, c2, c3, c4
```

```python
df.loc[c1, "Type"] = "[0, 90.6]"
df.loc[c2, "Type"] = "[90.6, 91.7]"
df.loc[c3, "Type"] = "[91.7, 93]"
df.loc[c4, "Type"] = "[93, 99]"
```

```python
df1 = df.rename(columns = {"Type":"Eff_Metric"})
```

```python
df1
```

```python
df2 = df1[df1['FOREHEART LINE P5 ZONE 1 TOP TEMPERATURE'] > 500]
#rslt_df = dataframe[dataframe['FOREHEART LINE P5 ZONE 1 TOP TEMPERATURE'] > 500]
#df_names = pd.read_csv("box_plots.csv")
```

```python
# li
li = ['CaO','Al2O3','Fe2O3','K2O','MgO','Na2O','SiO2','TiO2','Density (g/cm3)','Total Cullet %','FOREHEART LINE P5 ZONE 1 TOP TEMPERATURE','FOREHEART LINE P5 ZONE 1 MIDDLE TEMPERATURE','FOREHEART LINE P5 ZONE 1 BOTTOM TEMPERATURE','FOREHEART LINE P5 ZONE 2 TEMPERATURE','FOREHEART LINE P5 ZONE 3 TEMPERATURE','P5_seed']
```

```python
for elem in li:
    #if(elem=="Name" or elem=="Date" or elem=="FGCode"):
    #    continue
    seaborn.boxplot(x="Eff_Metric",
                y=elem,
                data=df2, order=["[0, 90.6]", "[90.6, 91.7]", "[91.7, 93]", "[93, 99]"])
    if (elem!= 'Density (g/cm3)'):
        plt.savefig("{}.png".format(elem))
        plt.clf()
    else:
        plt.savefig("Density.png")
        plt.clf()
```

```python
new_vars = ['FOREHEART LINE P5 ZONE 4 TEMPERATURE','FOREHEART LINE P5 ZONE 5 TEMPERATURE','Barrier Boosting (kWh)','Boosting (kWh)','Barrier Boosting Per ton (kWh/ton)','FO density (Specific Gravity)','FO L/ton']
for elem in new_vars:
    seaborn.boxplot(x="Eff_Metric",
                y=elem,
                data=df2, order=["[0, 90.6]", "[90.6, 91.7]", "[91.7, 93]", "[93, 99]"])
    a = str(elem).replace('/','_')
    plt.savefig(a)
    plt.show()
#     if (elem!= 'Density (g/cm3)'):
#         plt.savefig("{}.png".format(elem))
#         plt.clf()
#     else:
#         plt.savefig("Density.png")
#         plt.clf()
```

```python
seaborn.boxplot(x="Eff_Metric",
                y="FO L/ton",
                data=df2, order=["[0, 90.6]", "[90.6, 91.7]", "[91.7, 93]", "[93, 99]"])
```
