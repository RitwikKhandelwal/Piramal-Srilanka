```python
df2 = pd.read_csv("Defect_Data.csv")
df2 = df2[df2["Line"]=="H02"]
df2= df2[["Job Name", "Start", "End", "Defect Type", "%", "percent"]]
df2
```

```python
df2['Start'] = df2.apply(lambda d: pd.date_range(d['Start'],
                                                   d['End'],
                                                   freq='1d')[:-1],
                           axis=1)
df2
```

```python
df2 = df2.explode('Start')
df2['End'] = df2['Start'] + pd.Timedelta('1d')
df2 = df2.reset_index()
df2 = df2.drop("index", axis=1)
df2.head(10)
```

```python
# int(df2["%"][2].split(" ")[0])
df2["Defect Count"] = [0]*len(df2)
df2["Sample Size"]  = [0]*len(df2)
```

```python
for ind in df2.index:
    df2["Defect Count"][ind] = int(df2["%"][ind].split(" ")[0])
    df2["Sample Size"][ind] = df2["Defect Count"][ind]*100/df2["percent"][ind]
```

```python
df2 = df2.drop(["%", "percent"], axis=1)
```
